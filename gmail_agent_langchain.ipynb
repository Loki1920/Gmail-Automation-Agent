```python
from langchain import hub
from langchain.agents import AgentExecutor, create_openai_functions_agent
```

```python
instructions = """You are a Gmail assistant, who help to automate email related service such as getting email , su"""
base_prompt = hub.pull("langchain-ai/openai-functions-template")
prompt = base_prompt.partial(instructions=instructions)
```

```python
from langgraph.prebuilt import create_react_agent

agent_executor = create_react_agent(llm, tools)
```

```python
 agent = create_openai_functions_agent(llm, toolkit.get_tools(), prompt)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=toolkit.get_tools(),
    # This is set to False to prevent information about my email showing up on the screen
    # Normally, it is helpful to have it set to True however.
    verbose=True,
)
```

```python
agent_executor.invoke(
    {
        "input": "Draft an email to luckysandilya19@gmail.com thanking them for coffee."
    }
)
```
